```python
def add(a,b):
    print(a+b)
def sub(a,b):
    print(a-b)
def mul(a,b):
    print(a*b)
def div(a,b):
    print(a/b)
def cal():
    a = float(input("Enter a = "))
    b = float(input("Enter b = "))
    o = input("Enter + - * /")
    if o == '+':
        add(a,b)
    elif o == '-':
        sub(a,b)
    elif o == '*':
        mul(a,b)
    elif o == '/':
        div(a,b)
    else:
        print("Invalid operator")
cal()
```

```python
#Function with argument and return type
def cal(l,b):
    a = l*b
    return(a)
l = int(input("Enter l = "))
b = int(input("Enter b = "))
h = int(input("Enter h = "))
ar = cal(l,b)
print("The area is = ",ar)
print("The volume is = ",ar*h)
```

```python
def cal(l,b,h):
    a = l*b
    v = a*h
    return a,v
l = int(input("Enter l = "))
b = int(input("Enter b = "))
h = int(input("Enter h = "))
ar = cal(l,b,h)
area,volume = ar
print("The area is = ",area)
print("The volume is = ",volume)
```

```python
def area(l,b):
    a = l*b
    return a
def volume(l,b,h):
    v = area(l,b)*h
    return v
l = int(input("Enter l = "))
b = int(input("Enter b = "))
h = int(input("Enter h = "))
print("The area is = ",area(l,b))
print("The volume is = ",volume(l,b,h))
```
